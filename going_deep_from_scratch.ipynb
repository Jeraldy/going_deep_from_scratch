### Data pre processing

```python
# Reshape the training and test examples 
X = X.reshape(X.shape[0], -1)/255
Y = Y.reshape((Y.shape[0],1))

print ("X's shape: " + str(X.shape))
print ("Y's shape: " + str(Y.shape))
```

## 3. Helper Functions

```python
def init_params(layer_dims):
    L = len(layers_dims)
    params = {}
    for l in range(1,L):
        params["W" + str(l)] = np.random.randn(layer_dims[l],layer_dims[l-1])*0.1
        params["b" + str(l)] = np.zeros((layer_dims[l], 1))
    return params

def foward_prop(A,W,b):
    Z = W.dot(A) + b
    A = 1/(1+np.exp(-Z))    # Sigmoid
    assert(Z.shape == (W.shape[0], A.shape[1]))
    return (Z,A)

def compute_cost(AL, Y):
    m = Y.shape[1]
    cost = (-1 / m) * np.sum(np.multiply(Y, np.log(AL)) + np.multiply(1 - Y, np.log(1 - AL)))
    cost = np.squeeze(cost)
    return cost

def sigmoid_backward(dA, Z):  
    s = 1/(1+np.exp(-Z))
    dZ = dA * s * (1-s)
    assert(dZ.shape == Z.shape)
    return dZ
```

## 4. Initializing Paramers

```python
layers_dims = [X.shape[1],7,1]
m = X.shape[0]

lr = 0.0075
epochs = 25000
params = init_params(layers_dims)
cache = {}
grads = {}
L = len(layers_dims)

X = X.T
Y = Y.T

costs = []
cache["A0"] = X
```

## 6. Traning a neural Network

```python
for i in range(epochs):
    # Foward Prop
    for l in range(1,L):
        ZL,AL = foward_prop(cache["A"+str(l-1)],params["W"+str(l)],params["b"+str(l)])
        cache["A"+str(l)] = AL
        cache["Z"+str(l)] = ZL

    # Back Prop for the Last Layer
    dAL = - (np.divide(Y, AL) - np.divide(1 - Y, 1 - AL))    
    dZ = sigmoid_backward(dAL, ZL) 
    
    grads["dZ"+str(L-1)] = dZ
    grads["dW"+str(L-1)] = (1/m)*np.dot(dZ,AL.T)
    grads["db"+str(L-1)] = (1/m)*np.sum(dZ,axis=1,keepdims=True)
    
    # Back Prop for the Sub-sequent Layers
    for l in reversed(range(1,L-1)): 
        dA = np.dot(params["W"+str(l+1)].T, grads["dZ"+str(l+1)] ) 
        dZ  = sigmoid_backward(dA, cache["Z"+str(l)])
      
        dW = np.dot(dZ, cache["A"+str(l-1)].T) / m
        db = np.sum(dZ, axis=1, keepdims=True) / m
      
        assert (dA.shape == cache["A"+str(l)].shape)
        assert (dW.shape == params["W"+str(l)].shape)
        assert (db.shape == params["b"+str(l)].shape)
        
        grads["dW"+str(l)] = dW
        grads["db"+str(l)] = db
        grads["dZ"+str(l)] = dZ
      
    # Gradient Descent
    for l in range(1,L):
        params["W"+str(l)] = params["W"+str(l)] - lr*grads["dW"+str(l)]
        params["b"+str(l)] = params["b"+str(l)] - lr*grads["db"+str(l)]
        
    # Loss
    cost = compute_cost(cache["A"+str(L-1)],Y) 
    if i%1000 == 0:
        costs.append(cost)
        print("================== Iteration:-",i," =====================")
        print("Loss = ",cost)
```

## 7. Model Accuracy

```python
def accuracy(X, y, params):
    m = X.shape[1]
    p = np.zeros((1, m),dtype=int)
    A = X
    
    # Forward propagation
    for l in range(1,L):
        _,A = foward_prop(A,params["W"+str(l)],params["b"+str(l)])

    # convert probas to 0/1 
    ions
    for i in range(0, A.shape[1]):
        if A[0,i] > 0.5:
            p[0,i] = 1
        else:
            p[0,i] = 0

    print("Accuracy: %s" % str(np.sum(p == y)/float(m)))
```

```python
accuracy(X, Y, params)
```
